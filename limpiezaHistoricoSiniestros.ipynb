### 3.7 Borrado de columnas no apropiadas para modelamiento

Para esto debemos tener a consideración lo siguiente:

1. Qué significa cada variable? 
2. El tipo de dato que representa es acorde a como Pandas lo detecta inicialmente?
3. La presencia de null values es indicio de que debería ser eliminada?

```python
colisiones.tail(20)
```

Teniendo en consideración las pautas anteriores, tomamos la decisión de eliminar las tres primeras columnas:
- X
- Y
- OBJECTID

Esto debido a que X e Y, al hacer referencia a coordenadas geográficas decimales en longitud y latitud, ya están representadas en las columnasLATITUD y LONGITUD.

Además, OBJECTID hace referencia a un identificador único, por lo cuál NO es necesario que lo tengamos en cuenta

```python
colisiones.columns
```

```python
colisiones.drop(columns=['X','Y','OBJECTID'],inplace=True)
```

```python
colisiones.tail(10)
```

### 3.8 Identificamos valores null y los tratamos de acuerdo al caso

```python
#Adicionalmente, debido a que en la columna 'PK_CALZADA' tenemos varios registros que aparecen como Nan, podemos reemplazarlos con 
#un valor más explicativo, ya qie no podemos borrar el registro
colisiones['PK_CALZADA'] = colisiones['PK_CALZADA'].fillna('SIN CALZADA')
colisiones['CIV'] = colisiones['CIV'].fillna(0)
colisiones['LOCALIDAD'] = colisiones['LOCALIDAD'].fillna('SIN INFORMACION')
colisiones.tail(10)
```

### 3.9. Tratamos la columna de FECHA:

Identificamos que inicialmente Pandas la toma como de tipo object: Debemos convertirla a formato fecha yt extraer la información de fecha y hora.

Consulta este enlace: https://stackoverflow.com/questions/62917882/convert-datetime64ns-utc-pandas-column-to-datetime

```python
#Adicionalmente nos damos cuenta que la columna de fechas está en formato string: Convirtámosla a fecha y hora
# Convert 'FECHA_HORA_ACC' series to datetime dtype
#colisiones['FECHA_HORA_ACC'] = pd.to_datetime(colisiones['FECHA_HORA_ACC']) # si se deja así, el dtype retorn datetime64[ns, UTC]. NO es correcto
#colisiones['FECHA_OCURRENCIA_ACC'] = pd.to_datetime(colisiones['FECHA_OCURRENCIA_ACC']) # si se deja así, el dtype retorn datetime64[ns, UTC]. NO es correcto
colisiones['FECHA_HORA_ACC'] = pd.to_datetime(colisiones.FECHA_HORA_ACC).dt.tz_localize(None) #Removermos la timezone
colisiones['FECHA_OCURRENCIA_ACC'] = pd.to_datetime(colisiones.FECHA_OCURRENCIA_ACC).dt.tz_localize(None) #Removermos la timezone
colisiones[['FECHA_HORA_ACC','FECHA_OCURRENCIA_ACC']]
```

```python
# Creamos dos nuevas Pandas Series
colisiones['FECHA_ACC'] = colisiones['FECHA_HORA_ACC'].dt.date
colisiones['HORA_ACC'] = colisiones['FECHA_HORA_ACC'].dt.time
#Dbeemos asegurarnos de que esas Pandas Series estén en formato datetime
```
